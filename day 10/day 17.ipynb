```python
import matplotlib.pyplot as plt
import numpy as np
```

# type of graph
1.plot-              line graph,between two parameters
2.bar-               bar graph of single parameter
3.hist-              histogram of a bunch of data i.e.frequency of single parameter
4.scatter-           dot graph between two graph
5.boxplot/ box-      sarcastical overviewis for box for a single param
                     mean, outliner, quantiles(25%,75%), IQR
                     (optional - catogrical data)
6.stem-              more like bar graph, lines instead of bars
7.violinplot-        violin in shape statistical overview graph spread of data
)

```python
x = np.linspace(0,5,11)
x1 = np.arange(1,11,3)
```

```python
x1
```

```python
y=x**3
```

```python
plt.plot(x,y)
#plt.show()
```

```python
y2=x**2
```

```python
plt.plot(x,y,x,y2)
#plt.show()
```

```python
plt.plot(x,y,x,y2)
plt.legend(['y=x**3', 'y2=x**2'])
#plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,y2)
plt.legend(['y=x**3', 'y2=x**2'])
#plt.show()
```

```python
fig=plt.figure(figsize=(10,10))
plt.plot(x,y, lw=3, marker='o', markersize=15, markerfacecolor="yellow")
plt.plot(x,y2, marker='v', markersize=11)
plt.legend(['y=x**3', 'y2=x**2'])
#plt.show()
```

```python
fig.save
```

```python
plt.bar(x,y)
plt.title('bar graph')
```

```python
plt.hist(x)
plt.title('histogram')
```

```python
plt.scatter(x,y)
plt.title('scatter')
plt.grid()
plt.xlabel('x -- >')
plt.ylabel('y -- <')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
plt.boxplot(x)
plt.title('box')
```

```python
plt.stem(x,y)
plt.title('stem')
```

```python
plt.violinplot(x)
plt.title('violinplot')
```

```python
plt.violinplot(y)
plt.title('violinplot')
```
